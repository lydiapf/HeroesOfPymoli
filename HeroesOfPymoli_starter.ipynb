## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
player_dem = purchase_data.loc[:, ["Gender", "SN", "Age"]].drop_duplicates()
total_genders = player_dem["Gender"].value_counts()
total_genders
gender_percentage = total_genders/unique_players
gender_percentage
```

```python
gen_dem = pd.DataFrame({"Total Count": total_genders,
                       "Percentage of Players": gender_percentage})
gen_dem["Percentage of Players"] = gen_dem["Percentage of Players"].map("{:,.2%}".format)
gen_dem
```
