## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Get rid of duplicates within Gender, SN, and Age categories 
player_dem = purchase_data.loc[:, ["Gender", "SN", "Age"]].drop_duplicates()

# Now count the clean data to get Gender info
total_genders = player_dem["Gender"].value_counts()
total_genders

# Divide by previously found unique players to get gender percentage
gender_percentage = total_genders/unique_players
gender_percentage
```

```python
gen_dem = pd.DataFrame({"Total Count": total_genders,
                       "Percentage of Players": gender_percentage})
gen_dem["Percentage of Players"] = gen_dem["Percentage of Players"].map("{:,.2%}".format)
gen_dem
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python


spender_count = purchase_data.groupby(["SN"]).count()["Price"]
spender_avg = purchase_data.groupby(["SN"]).mean()["Price"]
spender_total = purchase_data.groupby(["SN"]).sum()["Price"]

# Create Data Frame
spender_data = pd.DataFrame({"Purchase Count": spender_count, 
                          "Average Purchase Price": spender_avg, 
                          "Total Purchase Value": spender_total})

# Format $ and sort by descending order
spender_data_format = spender_data.sort_values("Total Purchase Value", ascending = False)
spender_data_format["Average Purchase Price"] = spender_data["Average Purchase Price"].map("${:,.2f}".format)
spender_data_format["Total Purchase Value"] = spender_data["Total Purchase Value"].map("${:,.2f}".format)

# Display df
spender_data_format.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group items by Item ID, then use count, sum, and unique functions to get usable info
# Str hard bracket 0 is used to get rid of weird bracket formatting in table
item_id = purchase_data.groupby(purchase_data['Item ID'])
unique_items = item_id['Item ID'].unique().str[0]
item_name = item_id['Item Name'].unique().str[0]
item_purchase_count = item_id['Age'].count()
item_price = item_id['Price'].unique().str[0]
item_purchase_total = (item_id['Price'].sum())

# Create Data Frame
item_summary =pd.DataFrame({'Item Name':item_name,
                'Purchase Count':item_purchase_count,
                'Item Price':item_price,
                'Total Purchase Value':item_purchase_total})

# Format $ and by descending order
item_summary_format = item_summary.sort_values('Purchase Count', ascending=False)
item_summary_format["Item Price"] = item_summary["Item Price"].map("${:,.2f}".format)
item_summary_format["Purchase Count"] = item_summary["Purchase Count"].map("${:,.2f}".format)
item_summary_format["Total Purchase Value"] = item_summary["Total Purchase Value"].map("${:,.2f}".format)

# Display Data Frame
item_summary_format.head()
```
